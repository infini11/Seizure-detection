## Librairies

```python
import os
import glob

import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.neighbors import LocalOutlierFactor
from sklearn.impute import SimpleImputer

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.stats import zscore
from bioinfokit.visuz import cluster
```

## Load dataset

```python
# Path to dir to save figures
figures_path = 'figures/'
# List of files
files_path = "../seizure_detection_pipeline/output/cons-v0_6/*/cons_PAT_*.csv"
# # Path to output files
res_path = "res/general/"

# Creation of a list of file path
patient_files = glob.glob(files_path, recursive = True)
df_files = pd.DataFrame(patient_files, columns = ['path'])


def extract_patient_id(path):
    return int(path.split('cons_PAT_')[1].split('_Annotations')[0])

def extract_session_id(path):
    return int(path.split('_s')[0].split('_')[-1])

def extract_segment_id(path):
    return int(path.split('_s')[1].split('.csv')[0])


df_files['patient_id'] = df_files['path'].apply(extract_patient_id)
df_files['session_id'] = df_files['path'].apply(extract_session_id)
df_files['segment_id'] = df_files['path'].apply(extract_segment_id)
```

### Imputation of missing values

```python
df_session_18_copy.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
X = df_session_18_copy.drop(['label'], axis=1)
Y = df_session_18_copy['label']
```

```python
np.isinf(X).sum()
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(X)

X_imputed = imp_mean.transform(X)
```

```python
X_imputed
```

### Outlier detection and treatment

```python
# Detection des outliers
LOF = LocalOutlierFactor() ##Creation de l'objet LocalOutlierFactor pour la detection des outliers
Y_pred = LOF.fit_predict(X_imputed) ## entrainement et prediction des outliers probables
X_score = LOF.negative_outlier_factor_  ## recuperation du score des outliers negatifs
```

```python
outlier_score = pd.DataFrame() # creation d'un dataframe vide
outlier_score['score'] = X_score # remplissage avec la variable score
outlier_score.head()
```

```python
X_imputed = pd.DataFrame(X_imputed, columns=X.columns)
```

```python
X_imputed
```

```python
X_imputed.columns
```

```python
# representation et affissage des outliers
radius = (X_score.max() - X_score) / (X_score.max() - X_score.min())
outlier_score["radius"] = radius
filt = outlier_score["score"] < -2.5
outlier_index = outlier_score[filt].index.tolist()
plt.figure(figsize = (12,8))
plt.scatter(X_imputed.iloc[outlier_index,0], X_imputed.iloc[outlier_index,1], 
            color = "red", s = 200, label = "Outliers" )
plt.scatter(df_session_18.iloc[:,1], df_session_18.iloc[:,2], color ="blue", s=3, label = "Data Points" )

plt.legend()
plt.show()
```

```python
## suppression des valeurs outliers
X_imputed = pd.DataFrame(X_imputed).drop(outlier_index)
Y = Y.drop(outlier_index).values
```

```python
plt.scatter(X_imputed.iloc[:,0], X_imputed.iloc[:,1], color ="k", s=3)
```

### PCA Analysis

```python
X_imputed = zscore(X_imputed)
```

```python
X_imputed
```

```python
pca = PCA(n_components=0.95)
pca_out = pca.fit(X_imputed)
principalComponents = pca.transform(X_imputed)
```

```python
fig, ax = plt.subplots()
xi = np.arange(1, 10, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 11, step=1)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

```python
PC = range(1, pca.n_components_+1)
plt.bar(PC, pca.explained_variance_ratio_, color='gold')
plt.xlabel('Principal Components')
plt.ylabel('Variance %')
plt.xticks(PC)

# Putting components in a dataframe for later
PCA_components = pd.DataFrame(principalComponents)
```

```python
plt.scatter(PCA_components[0], PCA_components[1], alpha=.3, color='gold')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
plt.scatter(PCA_components[1], PCA_components[2], alpha=.3, color='gold')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
plt.scatter(PCA_components[2], PCA_components[3], alpha=.3, color='gold')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
def biplot(score, coeff , y):

    xs = score[:,0] # projection on PC1
    ys = score[:,1] # projection on PC2
    n = coeff.shape[0] # number of variables
    plt.figure(figsize=(10,8), dpi=100)
    classes = np.unique(y)
    colors = ['b','r','y']
    markers=['o','^','x']
    for s,l in enumerate(classes):
        plt.scatter(xs[y==l],ys[y==l], c = colors[s], marker=markers[s]) # color based on group
    for i in range(n):
        #plot as arrows the variable scores (each variable has a score for PC1 and one for PC2)
        plt.arrow(0, 0, coeff[i,0], coeff[i,1], color = 'k', alpha = 0.9,linestyle = '-',linewidth = 1.5, overhang=0.2)
        plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'k', ha = 'center', va = 'center',fontsize=10)

    plt.xlabel("PC{}".format(1), size=14)
    plt.ylabel("PC{}".format(2), size=14)
    limx= int(xs.max()) + 1
    limy= int(ys.max()) + 1
    plt.xlim([-limx,limx])
    plt.ylim([-limy,limy])
    plt.grid()
    plt.tick_params(axis='both', which='both', labelsize=14)
```

```python
biplot(X_imputed[:,0:2], np.transpose(pca.components_[0:2, :]), Y)
plt.show()
```
